### Remove emoji in column "รายละอียด"

```python
import re
```

```python
def remove_emojis(text):
    emoji_pattern = re.compile("["
                                u"\U0001F600-\U0001F64F"  # จาก emoticons
                           u"\U0001F300-\U0001F5FF"  # จาก symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # จาก transport & map symbols
                           u"\U0001F700-\U0001F77F"  # จาก alchemical symbols
                           u"\U0001F780-\U0001F7FF"  # จาก Geometric Shapes Extended
                           u"\U0001F800-\U0001F8FF"  # จาก Supplemental Arrows-C
                           u"\U0001F900-\U0001F9FF"  # จาก Supplemental Symbols and Pictographs
                           u"\U0001FA00-\U0001FA6F"  # จาก Chess Symbols
                           u"\U0001FA70-\U0001FAFF"  # จาก Symbols and Pictographs Extended-A
                           u"\U0001F004"             # จาก mahjong red dragon
                           u"\U0001F0CF"             # จาก playing cards
                           u"\U0001F18E"             # จาก negative squared ab
                           u"\U0001F191-\U0001F19A"  # จาก Squared ID
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002500-\U00002BEF"  # chinese char
                               "]+", flags=re.UNICODE)
    return re.sub(emoji_pattern, '', text)

# Remove emojis in the specified column
df['รายละเอียด'] = df['รายละเอียด'].apply(remove_emojis)
```

```python
from pythainlp.corpus import thai_stopwords

# Get the list of Thai stopwords
stop_words = thai_stopwords()

# Function to remove stopwords
def remove_stopwords(tokens):
    return [token for token in tokens if token not in stop_words]

# Apply the remove_stopwords function to the 'Normalize_Tokens2' column
df['stop_words'] = df['Normalize_Tokens_respace'].apply(remove_stopwords)
df['stop_words']
```

# POS_TAGER

```python
from pythainlp.tag import pos_tag
```

```python
# สร้างฟังก์ชันสำหรับทำ POS Tagging
def pos_tagging(tokens):
    tags = pos_tag(tokens)
    return tags
```

```python
# ใช้ DataFrame ที่มีคอลัมน์ "Tokens" และสร้างคอลัมน์ใหม่ "POS_Tags"
df['POS_Tags'] = df['stop_words'].apply(pos_tagging)
```

### นับความถี่ของคำหรือโทเค็น
นับความถี่ของแต่ละคำหรือโทเค็นในข้อมูล POS_Tags

```python
from collections import defaultdict

# สร้างโครงสร้างข้อมูลเพื่อเก็บความถี่ของคำหรือวลีที่ผ่านการโทเค็นและลบ stopwords
word_frequency = defaultdict(int)

# วนลูปผ่านข้อมูลคอลัมน์ 'stop_words' ใน DataFrame
for stop_words in df['POS_Tags']:
    for word in stop_words:
        word_frequency[word] += 1

# เรียงลำดับคำหรือวลีตามความถี่
sorted_words = sorted(word_frequency.items(), key=lambda x: x[1], reverse=True)
sorted_words
```

# TF-IDF

```python
from collections import Counter
import pandas as pd
import math
```

```python
# Flatten the list of tokens
flat_tokens = [token for tokens in df['POS_Tags'] for token in tokens]

# Count term frequency
tf_counter = Counter(flat_tokens)

# Calculate the total number of documents
total_documents = len(df)

# Calculate IDF for each term
idf_values = {}
for term, tf in tf_counter.items():
    idf = math.log10(total_documents / (1 + tf))  # Using base 10 logarithm for IDF
    idf_values[term] = idf

# Add IDF values to the tf_df DataFrame
tf_df = pd.DataFrame(tf_counter.items(), columns=['Word', 'TF'])
tf_df['IDF'] = tf_df['Word'].map(idf_values)

# Calculate TF-IDF and add it to the DataFrame
tf_df['TF-IDF'] = tf_df['TF'] * tf_df['IDF']

# Sort by TF-IDF in descending order
tf_df = tf_df.sort_values(by='TF-IDF', ascending=False)

# Display the DataFrame with TF, IDF, and TF-IDF
tf_df
```
