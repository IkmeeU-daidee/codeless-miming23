### Remove emoji in column "‡∏£‡∏≤‡∏¢‡∏•‡∏∞‡∏≠‡∏µ‡∏¢‡∏î"

```python
import re
```

```python
def remove_emojis(text):
    emoji_pattern = re.compile("["
                                u"\U0001F600-\U0001F64F"  # ‡∏à‡∏≤‡∏Å emoticons
                           u"\U0001F300-\U0001F5FF"  # ‡∏à‡∏≤‡∏Å symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # ‡∏à‡∏≤‡∏Å transport & map symbols
                           u"\U0001F700-\U0001F77F"  # ‡∏à‡∏≤‡∏Å alchemical symbols
                           u"\U0001F780-\U0001F7FF"  # ‡∏à‡∏≤‡∏Å Geometric Shapes Extended
                           u"\U0001F800-\U0001F8FF"  # ‡∏à‡∏≤‡∏Å Supplemental Arrows-C
                           u"\U0001F900-\U0001F9FF"  # ‡∏à‡∏≤‡∏Å Supplemental Symbols and Pictographs
                           u"\U0001FA00-\U0001FA6F"  # ‡∏à‡∏≤‡∏Å Chess Symbols
                           u"\U0001FA70-\U0001FAFF"  # ‡∏à‡∏≤‡∏Å Symbols and Pictographs Extended-A
                           u"\U0001F004"             # ‡∏à‡∏≤‡∏Å mahjong red dragon
                           u"\U0001F0CF"             # ‡∏à‡∏≤‡∏Å playing cards
                           u"\U0001F18E"             # ‡∏à‡∏≤‡∏Å negative squared ab
                           u"\U0001F191-\U0001F19A"  # ‡∏à‡∏≤‡∏Å Squared ID
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002500-\U00002BEF"  # chinese char
                               "]+", flags=re.UNICODE)
    return re.sub(emoji_pattern, '', text)

# Remove emojis in the specified column
df['‡∏£‡∏≤‡∏¢‡∏•‡∏∞‡πÄ‡∏≠‡∏µ‡∏¢‡∏î'] = df['‡∏£‡∏≤‡∏¢‡∏•‡∏∞‡πÄ‡∏≠‡∏µ‡∏¢‡∏î'].apply(remove_emojis)
```

```python
from pythainlp.tag import pos_tag
```

```python
from collections import defaultdict

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏Å‡πá‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏≥‡∏´‡∏£‡∏∑‡∏≠‡∏ß‡∏•‡∏µ‡∏ó‡∏µ‡πà‡∏ú‡πà‡∏≤‡∏ô‡∏Å‡∏≤‡∏£‡πÇ‡∏ó‡πÄ‡∏Ñ‡πá‡∏ô‡πÅ‡∏•‡∏∞‡∏•‡∏ö stopwords
word_frequency = defaultdict(int)

# ‡∏ß‡∏ô‡∏•‡∏π‡∏õ‡∏ú‡πà‡∏≤‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'stop_words' ‡πÉ‡∏ô DataFrame
for stop_words in df['POS_Tags']:
    for word in stop_words:
        word_frequency[word] += 1

# ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏Ñ‡∏≥‡∏´‡∏£‡∏∑‡∏≠‡∏ß‡∏•‡∏µ‡∏ï‡∏≤‡∏°‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà
sorted_words = sorted(word_frequency.items(), key=lambda x: x[1], reverse=True)
sorted_words
```

# TF-IDF

```python
from collections import Counter
import pandas as pd
import math
```

```python
# Flatten the list of tokens
flat_tokens = [token for tokens in df['POS_Tags'] for token in tokens]

# Count term frequency
tf_counter = Counter(flat_tokens)

# Calculate the total number of documents
total_documents = len(df)

# Calculate IDF for each term
idf_values = {}
for term, tf in tf_counter.items():
    idf = math.log10(total_documents / (1 + tf))  # Using base 10 logarithm for IDF
    idf_values[term] = idf

# Add IDF values to the tf_df DataFrame
tf_df = pd.DataFrame(tf_counter.items(), columns=['Word', 'TF'])
tf_df['IDF'] = tf_df['Word'].map(idf_values)

# Calculate TF-IDF and add it to the DataFrame
tf_df['TF-IDF'] = tf_df['TF'] * tf_df['IDF']

# Sort by TF-IDF in descending order
tf_df = tf_df.sort_values(by='TF-IDF', ascending=False)

# Display the DataFrame with TF, IDF, and TF-IDF
tf_df
```
